```python
def scale_curve(unscaled_arr):
    len(unscaled_arr)
    max_val=max(unscaled_arr) 
    #scaling the values between 0 to 100
    return unscaled_arr*100/max_val

cuts_orig_scaled=scale_curve(np.asarray(cuts_orig))
motion_orig_scaled=scale_curve(np.asarray(motion_orig))
sound_orig_scaled=scale_curve(np.asarray(sound_orig))

plt.plot(cuts_orig_scaled,'r')
plt.plot(motion_orig_scaled,'g')
plt.plot(sound_orig_scaled,'b')
```

```python
#trying to smoothen out the average curve obtained
#by the orignal curves , then smoothing it out

avg_orig=(cuts_orig_scaled+motion_orig_scaled+sound_orig_scaled)/3

def smooth_func(length,alpha):
    #kaiser_func=np.kaiser(len(avg_orig)/500,10)
    return np.kaiser(length,alpha)

kaiserfn=smooth_func(len(avg_orig)/50,10)
avg_orig_smooth = np.convolve(kaiserfn/kaiserfn.sum(),avg_orig,mode='same')
plt.plot(avg_orig_smooth)
```

```python
#averaging the curves after smoothing them out

cuts_smooth= np.convolve(kaiser_func/kaiser_func.sum(),cuts_orig,mode='same')
motion_smooth = np.convolve(kaiser_func/kaiser_func.sum(),motion_orig,mode='same')
sound_smooth = np.convolve(kaiser_func/kaiser_func.sum(),sound_orig,mode='same')

cuts_scaled_smooth = scale_curve(cuts_smooth)
motion_scaled_smooth = scale_curve(motion_smooth)
sound_scaled_smooth = scale_curve(sound_smooth)

plt.plot(cuts_scaled_smooth,'r')
plt.plot(motion_scaled_smooth,'g')
plt.plot(sound_scaled_smooth,'b')
plt.plot(kaiser_func)

avg= (cuts_scaled_smooth+motion_scaled_smooth+sound_scaled_smooth)/3
plt.plot(avg)
avg_smooth = np.convolve(kaiser_func/kaiser_func.sum(),avg ,mode='same')
```

```python
#multiplying orignal curves

mult_orig= [cuts_orig_scaled[k]*motion_orig_scaled[k]*sound_orig_scaled[k] for k in range(0,len(cuts_orig_scaled))]

kaiserfn1=smooth_func(len(avg_orig)/50,10)
mult_orig_smooth = np.convolve(kaiserfn1/kaiserfn1.sum(),mult_orig ,mode='same')
#plt.plot(mult_orig)
plt.plot(mult_orig_smooth)
```

```python
#multiplying smooth curves
mult_of_smooth= [cuts_scaled_smooth[k]*motion_scaled_smooth[k]*sound_scaled_smooth[k] for k in range(0,len(cuts_scaled_smooth))]
mult_smooth = np.convolve(kaiserfn1/kaiserfn1.sum(),mult_of_smooth ,mode='same')
plt.plot(mult_smooth)
```

```python
#filter peaks method
def get_filtered_peaks(smooth_curve):
    #find peaks from the curve
    increases = np.diff(smooth_curve)[:-1]>=0
    decreases = np.diff(smooth_curve)[1:]<=0
    peaks_position = ((increases * decreases).nonzero()[0])
    peaks_value = smooth_curve[peaks_position]
    peaks_position = peaks_position[peaks_value>np.percentile(peaks_value,90)]
    
    #filter two close (100 frames apart) peaks
    final_peaks_position=[peaks_position[0]]
    for fr_num in peaks_position:
        if (fr_num - final_peaks_position[-1]) < 200:
            if smooth_curve[fr_num] > smooth_curve[final_peaks_position[-1]]:
                final_peaks_position[-1] = fr_num
        else:
            final_peaks_position.append(fr_num)
    return final_peaks_position

print 'cuts_scaled_smooth', get_filtered_peaks(cuts_scaled_smooth) ,'\n'
print 'motion_scaled_smooth', get_filtered_peaks(motion_scaled_smooth) ,'\n'
print 'sound_scaled_smooth', get_filtered_peaks(sound_scaled_smooth) ,'\n'
print 'avg_orig_smooth_peaks= ', get_filtered_peaks(avg_orig_smooth) ,'\n'
print'avg_smooth_peaks= ', get_filtered_peaks(avg_smooth) ,'\n'
print ' mult_orig_smooth = ', get_filtered_peaks(mult_orig_smooth) ,'\n'
print ' mult_smooth = ', get_filtered_peaks(mult_smooth) ,'\n'
```
